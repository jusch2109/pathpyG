# Pathpy 2.0

```python
ks = range(1,10)
times = []
for k in ks:
    start = time.time() 
    paths = pp.Paths.read_file("tube_paths_train.ngram", max_subpath_length=k)
    g2 = pp.HigherOrderNetwork(paths, k=k)
    print(g2)
    elapsed_pp = time.time()-start
    times.append(elapsed_pp)
plt.plot(ks, times)
```

# pathpyG

```python
ks = range(1,10)
times_new = []
for k in ks:
    start = time.time()
    p = GlobalPathStorage.from_csv('tube_paths_train.ngram')
    gk = HigherOrderGraph(p, order=k, path_freq='path_freq')
    print(gk)
    print('---')
    elapsed_new = time.time()-start
    times_new.append(elapsed_new)
```

```python
plt.plot(ks, times, label='pathpy2')
plt.plot(ks, times_new, label='pathpyG prototype')
plt.xlabel('order')
plt.grid()
plt.ylabel('time [s]')
plt.legend()
```

```python
plt.plot(ks, times, label='pathpy2')
plt.plot(ks, times_new, label='pathpyG prototype')
plt.xlabel('order')
plt.ylabel('time [s]')
plt.legend()
plt.grid()
plt.yscale('log')
plt.ylim(1,100)
```
